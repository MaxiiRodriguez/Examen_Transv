```python
import numpy as np

ubicaciones = np.zeros(100)
Platinum = 120000
Gold = 80000
Silver = 50000
asistentes = {}


def comprar_entradas():
    cantidad = int(input("Ingrese la cantidad de entradas que desea comprar (Debe ser entre 1 a 3): "))
    while cantidad < 1 or cantidad > 3:
        print("Cantidad ingresada inválida. Intente nuevamente por favor.")
        cantidad = int(input("Ingrese la cantidad de entradas que desea comprar (Debe ser entre 1 a 3): "))
#return()

    mostrar_ubicaciones()

    for _ in range(cantidad):
        ubicacion = int(input("Ingrese la ubicación de asiento que desea, según la imagen referencial: "))
        while ubicacion < 1 or ubicacion > 100 or ubicaciones[ubicacion - 1] != 0:
            if ubicacion < 1 or ubicacion > 100:
                print("Ubicación seleccionada es inválida. Intente nuevamente.")
            else:
                print("Ubicación no disponible, en estos momentos. *Ya fue vendida*. Intente nuevamente.")
            ubicacion = int(input("Ingrese la ubicación de asiento que desea, según la imagen referencial: "))

        ubicaciones[ubicacion - 1] = 1

        run = input("Ingrese el RUN del asistente (Debe ser 8 u 9 dígitos, sin puntos ni guión y DV): ")
        while len(run) < 8 or len(run) > 9:
            print("RUN inválido. (Debe ser igual a 8 dígitos, sin puntos ni guión y DV)")
            run = input("Ingrese el RUN del asistente (Debe ser igual a 8 dígitos, sin puntos ni guión y DV): ")

        asistentes[ubicacion] = run

    print("Operación realizada correctamente.")


def mostrar_ubicaciones():
    print("\nEstado actual de la venta de entradas:\n")
    print("|===========================|")
    print("|*********ESCENARIO*********|\n")
    for i in range(10):
        for j in range(10):
            index = i * 10 + j
            if ubicaciones[index] == 0:
                estado = str(index + 1).rjust(2)
            else:
                estado = "X"
            print(f"{estado} ", end="")
        print()

    print("\n\nRecuerde: ")
    print("Los precios de las entradas son los siguientes:\n")
    print("|-------------|-----------------|------------------------------|")
    print("|TIPO ENTRADA |   VALOR         |   CONDICION DE ASIENTOS      |")
    print("|-------------|-----------------|------------------------------|")
    print("|Platinum     |   $120.000.-    |   (Asientos del 1 al 20)     |")
    print("|Gold         |   $80.000.-     |   (Asientos del 21 al 50)    |")
    print("|Silver       |   $50.000.-     |   (Asientos del 51 al 100)   |")
    print("|-------------|-----------------|------------------------------|\n\n")
  #return()


def ver_listado_asistentes():
    print("Listado de asistentes:")
    for nombre in asistentes:
        run = asistentes[nombre]
        print("Ubicación comprada: ",nombre,"\n"
        "run de asistente: ",run)
  #return()


def mostrar_ganancias_totales():
    vendidas = np.nonzero(ubicaciones)[0]
    total = sum(Platinum
      if i < 20
      else Gold
      if i < 50
      else Silver
      for i in vendidas)
    print("Ganancias totales: $",total)
#return()


def salir():
    print(
    "\n\n"
    "wwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwww\n"
    "     Gracias por preferirnos       \n"
    "     No olvides visitarnos en Creativos.cl   \n\n"
    "     Maximiliano Rodríguez, 11-07-2023\n"
    "wwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwwww\n")
#return()


while True:
    print("\n--------------- Menú ---------------\n")
    print("1- Comprar entradas")
    print("2- Mostrar ubicaciones disponibles")
    print("3- Ver listado de asistentes")
    print("4- Mostrar ganancias totales")
    print("5- Salir\n")
    print("Concierto VIP de Michael Jam, exclusivo!, solo para 100 asistentes\n"
    "\n*************** CUPOS LIMITADOS ***************")

    opcion = input("\nSeleccione una opción por favor: ")

    if opcion == "1":
        comprar_entradas()
    elif opcion == "2":
        mostrar_ubicaciones()
    elif opcion == "3":
        ver_listado_asistentes()
    elif opcion == "4":
        mostrar_ganancias_totales()
    elif opcion == "5":
        salir()
    else:
        print("Opción seleccionada es inválida. Intente nuevamente por favor.")
```
